```python
import os
import re
import shutil
path = r"C:\Users\lizhi\Desktop\Guigu/"
```

```python
L = []
import re
regex = r"(.*)(第)(\d{2,3})(封.)(.*)"
for mp3 in os.listdir(path):
    if "第" in mp3:
        origin = mp3
        mp3 = mp3.replace('丨', '')
        mp3 = mp3.replace('-', '')
        m = re.match(regex, mp3)
        if m:
            num = m.group(3)
            if len(num) < 3 :
                num = '0' + num
            print (num + ' ' + m.group(5))
            #shutil.copy(path+origin, num + ' ' + m.group(5))
            os.remove(path+origin)
```
